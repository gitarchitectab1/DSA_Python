```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

fig = plt.figure()
ax = Axes3D(fig, auto_add_to_figure=False)
fig.add_axes(ax)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.viridis)

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("StudentsPerformance.csv")
```

```python
df
```

```python
df.describe()
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.boxplot(data=df,x='parental level of education',y='math score',hue='test preparation course')
plt.legend(bbox_to_anchor=(1.05,0.95))
#plt.text('test preparation course',1,0.95)
plt.show()
```

```python
plt.figure(figsize=(5,8),dpi=200)
sns.violinplot(data=df,y='parental level of education',x='math score',hue='test preparation course')
plt.legend(bbox_to_anchor=(1.05,0.95))
#plt.text('test preparation course',1,0.95)
plt.show()
```

```python
plt.figure(figsize=(5,8),dpi=200)
sns.violinplot(data=df,y='parental level of education',x='math score',hue='test preparation course',split=True,inner=None)
plt.legend(bbox_to_anchor=(1.05,0.95))
#plt.text('test preparation course',1,0.95)
plt.show()
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.boxenplot(data=df,x='parental level of education',y='math score',hue='test preparation course')
plt.legend(bbox_to_anchor=(1.05,0.95))
#plt.text('test preparation course',1,0.95)
plt.show()
```

```python
plt.figure(figsize=(5,10),dpi=200)
sns.swarmplot(data=df,y='parental level of education',x='math score',hue='test preparation course',dodge=True)
plt.legend(bbox_to_anchor=(1.05,0.95))
#plt.text('test preparation course',1,0.95)
plt.show()
```

```python
plt.figure(figsize=(5,10),dpi=100)
sns.jointplot(data=df,y='reading score',x='math score',kind='hex')
#plt.legend(bbox_to_anchor=(1.05,0.95))
#plt.text('test preparation course',1,0.95)
plt.show()
```
